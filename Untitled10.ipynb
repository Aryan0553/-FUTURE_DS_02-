```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/newexcel2/marketing_and_product_performance.csv")
```

```python
df.describe()
```

```python
#missing values
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#cleaning data
df.fillna(0, inplace=True)
```

```python
#CTR (if impressions not available, use Click Efficiency)
df['Click_Efficiency'] = df['Clicks'] / df['Budget']
```

```python
#Conversion rate
df['Conversion_Rate'] = (df['Conversions'] / df['Clicks']) * 100
```

```python
#ROI (Recalculated for validation)
df['ROI_Calculated'] = ((df['Revenue_Generated'] - df['Budget']) / df['Budget']) * 100
```

```python
#Revenue per Conversion
df['Revenue_per_Conversion'] = df['Revenue_Generated'] / df['Conversions']
```

```python
#Campaign Performance Analysis (Code)
#Top Campaigns by ROI
df.groupby('Campaign_ID')['ROI_Calculated'].mean().sort_values(ascending=False).head()
```

```python
#Budget vs Revenue
df[['Budget','Revenue_Generated']].corr()
```

```python
#Subscription Tier Analysis (Code)
df.groupby('Subscription_Tier')[['Revenue_Generated','ROI_Calculated','Conversion_Rate']].mean()
```

```python
#Budget vs Revenue
plt.figure(figsize=(8,5))
sns.scatterplot(x='Budget', y='Revenue_Generated', data=df)
plt.title("Budget vs Revenue")
plt.show()
```

```python
#ROI Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['ROI_Calculated'], bins=30, kde=True)
plt.title("ROI Distribution")
plt.show()
```

```python
#Conversion Rate by Subscription Tier
plt.figure(figsize=(8,5))
sns.barplot(x='Subscription_Tier', y='Conversion_Rate', data=df)
plt.title("Conversion Rate by Subscription Tier")
plt.show()
```

```python
best_selling_product = (
    df.groupby('Product_ID')['Conversions']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

best_selling_product
```

```python
best_selling_product.plot(kind='bar', figsize=(8,5), title='Top 10 Best-Selling Products')
plt.ylabel("Total Conversions")
plt.show()
```

```python
sales_trend = df.groupby(df.index)['Conversions'].sum()

plt.figure(figsize=(8,5))
plt.plot(sales_trend)
plt.title("Sales Trend Over Records")
plt.xlabel("Record Index")
plt.ylabel("Conversions")
plt.show()
```

```python
revenue_trend = df.groupby(df.index)['Revenue_Generated'].sum()

plt.figure(figsize=(8,5))
plt.plot(revenue_trend, color='green')
plt.title("Revenue Trend")
plt.xlabel("Record Index")
plt.ylabel("Revenue")
plt.show()
```
